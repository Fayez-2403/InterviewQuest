```python
import csv

filename = "output_dbms_ngram_without_prepro_len_2.txt"

with open(filename, "r") as f, open("table_output_for_dbms_ngram_len2_without_prepro.csv", "w", newline="") as output_file:
    writer = csv.writer(output_file)
    writer.writerow(["Text", "Similarities"])

    lines = f.readlines()
    for i in range(len(lines)):
        line = lines[i].strip()
        if line.startswith("Before prepro :  "):
            text = line.strip("Before prepro :  ")
            sim_line = lines[i+2].strip()
            similarities = sim_line.split("similarities are [")[1].split("]")[0].split()
            writer.writerow([text, similarities])
```
