```python
from numba.core.errors import NumbaDeprecationWarning, NumbaPendingDeprecationWarning
import warnings

warnings.simplefilter('ignore', category=NumbaDeprecationWarning)
warnings.simplefilter('ignore', category=NumbaPendingDeprecationWarning)

import os
from bertopic import BERTopic
import pandas as pd


model_merged_ngram = BERTopic.load("../model_merged_ngram_no_len")

tp = model_merged_ngram.get_topic_info()

# Split the 'Bucket' column by underscores and join the parts with commas
tp['Bucket'] = tp['Name'].str.split('_').str[1:].str.join(',')

data = {
    'Bucket Number': tp['Topic'],
    'Bucket': tp['Bucket']
}

df = pd.DataFrame.from_dict(data)

df.to_csv('net_bucket.csv', index=False)
```

```python
import os
from bertopic import BERTopic

os.chdir("..")  # Assuming you've changed to the parent directory where the model is located
model_merged_ngram = BERTopic.load("model_merged_ngram_no_len")

# Define a function to print all the words in a topic
def print_all_topic_words(model, topic_id):
    words_with_weights = model.get_topic(topic_id)
    
    # Check if the topic is not empty
    if words_with_weights:
        words = [word for word, _ in words_with_weights]
        print(f"Topic {topic_id}: {', '.join(words)}")
    else:
        print(f"Topic {topic_id}: Empty Topic")

# Iterate over each topic, including -1, and print all the words
num_topics = len(model_merged_ngram.get_topic_freq().index)
for topic_id in range(-1, num_topics - 1):  # Adjust the range to include -1
    print_all_topic_words(model_merged_ngram, topic_id)
```

```python
#8 words
import os
from bertopic import BERTopic
import pandas as pd

os.chdir("..")  # Assuming you've changed to the parent directory where the model is located
model_merged_ngram = BERTopic.load("model_merged_ngram_no_len")
os.chdir("mapping_topic")
# Create an empty DataFrame to store the topic data
topic_data = pd.DataFrame(columns=["Bucket Number", "Bucket"])

# Define a function to add topic data to the DataFrame
def add_topic_data(model, topic_id, dataframe, max_words=8):
    words_with_weights = model.get_topic(topic_id)
    if words_with_weights:
        words = [word for word, _ in words_with_weights[:max_words]]  # Limit to the first 8 words
        topic_data.loc[topic_id] = [topic_id, ", ".join(words)]  # Separating words with commas

# Iterate over each topic, including -1, and add the topic data to the DataFrame
num_topics = len(model_merged_ngram.get_topic_freq().index)
for topic_id in range(-1, num_topics - 1):  # Adjust the range to include -1
    add_topic_data(model_merged_ngram, topic_id, topic_data)

# Save the DataFrame to a CSV file
topic_data.to_csv("net_bucket_new8.csv", index=False)
```
