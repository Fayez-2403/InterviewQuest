```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('compare_graph/os_compare_sorted.csv')
df = df.drop(df.index[-1])

x = df['difference'] 
y = df['question'].apply(lambda x: len(str(x).split()))

# Set the figure size to increase the width of the plot
plt.figure(figsize=(40, 10))  # Adjust the width (10) and height (6) as needed

plt.plot(x, y, marker='o',markerfacecolor='red', markersize=10)
plt.show()
```

```python
x = df['merged_ngram_probab'] 
y = df['question_word_count'] = df['question'].apply(lambda x: len(str(x).split()))

# Set the figure size to increase the width of the plot
plt.figure(figsize=(40, 10))  # Adjust the width (10) and height (6) as needed

plt.scatter(x, y)
plt.show()
```

```python
x = df['merged_ngram_probab'] 
y = df['question'].apply(lambda x: len(str(x).split()))

# Get the indices that would sort y in ascending order
sorted_indices = np.argsort(y)

# Apply the sorted indices to both x and y
x_sorted = x[sorted_indices]
y_sorted = y[sorted_indices]
# print(x_sorted)
# print(y_sorted)
plt.figure(figsize=(40, 30))

plt.plot(y_sorted,x_sorted, marker='o',markerfacecolor='red', markersize=10)
plt.show()
```
